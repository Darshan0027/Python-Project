```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
df=pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
#total number od rows and columns
df.shape
```

```python
#quickly viewed the top rows of your dataset that provides insight into the structure and values without printing the entire dataset
df.head(10)
```

```python
#concise summary of the data set using info()

df.info()
```

```python
#dropped unnecessary column using drop

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#Checked null values in each column
pd.isnull(df).sum()
```

```python
#dropped the null values
df.dropna(inplace=True)
```

```python
#changed the datatype

df['Amount']=df['Amount'].astype('int')
```

```python
#Checked for the changed datatype

df['Amount'].dtypes
```

```python
df.columns
```

```python
#renamd the column
df.rename(columns={'Gender':'Gender-M/F'})
```

# **EXPLORATORY DATA ANALYSIS**

**Gender**

```python
df.columns
```

```python
#No. of Males and Females

ax=sns.countplot(x='Gender',data=df,)
for bars in ax.containers:
  ax.bar_label(bars)
  plt.show()
```

***From the above plot we can conclude that most of the buyers are Females***

```python
Sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False  )
ax=sns.barplot(x='Gender',y='Amount',data=Sales_gen,palette='Set1')
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
```

*** From the above 2 plots we can conclude that Most of the buyers are Females and Purchasing power of females are also high***

```python
ax=sns.countplot(x='Age Group',data= df,hue = 'Gender',edgecolor='black')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Total amount vs Age group

Sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=Sales_age,palette='Paired',edgecolor='black')
plt.show()
```

**From the above 2 graph,we can see that most of the buyers are between 26-35 and most of them are females**

# **STATE**

```python
df.columns
```

```python
#Total Number of orders from top 10 states

Sales_state=df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=Sales_state,palette='Paired',edgecolor='black')
plt.show()
#
```

```python


#Total amount spent-Top 10 states

Sales_state=df.groupby(['State'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=Sales_state,palette='Paired',edgecolor='black')
plt.show()
```

***From the above graph we can draw insights that most of the orders and total amount spent are from Uttar Pradesh,Maharashtra and Karnataka***

# **Marital Status**

```python
ax=sns.countplot(data=df,x='Marital_Status',hue='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_MG=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data= Sales_MG )
```

***From the above 2 graph we can conclude that most of the buyers are married women***

# **Occupation**

```python
ax=sns.countplot(data=df,x='Occupation',hue='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_Oc=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=Sales_Oc,hue='Occupation')
```

***From the above graph we can conclude that most of the buyers are from IT,healthcare,Aviation***

# **Product category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x='Product_Category',hue='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_Pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=Sales_Pc,hue='Product_Category')
```

***From the above graphs we can conclude that most sold products are Food,clothing and electronics by amount ***

```python
Sales_PID=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=Sales_PID,hue='Product_ID')
```

***Above graph shows which Product_category has been sold more***

# **Insights**

### ***Married Women from age group 26-35 from UP,Maharashtra,Karnataka working in IT,Healthcare and aviation are more likely buy products from Food Clothing and Electronic category***
